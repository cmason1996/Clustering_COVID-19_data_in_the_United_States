```python
# Import dependencies
import pandas as pd
```

```python
# Read in the csv file
counties_df = pd.read_csv('../counties.timeseries.csv')
counties_df
```

```python
pd.set_option('display.max_rows', 70)
counties_df.dtypes
```

```python
# Get the length of each column
counties_df.count()
```

```python
# Create the list of columns to drop
columns_to_drop = ['country', 'lat', 'long', 'actuals.positiveTests', 'actuals.negativeTests', 
                   'actuals.contactTracers', 'unused1', 'unused2', 'actuals.vaccinesDistributed', 
                   'metrics.testPositivityRatio', 'metrics.testPositivityRatioDetails', 'metrics.caseDensity', 
                   'metrics.contactTracerCapacityRatio', 'metrics.infectionRate', 'metrics.infectionRateCI90', 
                   'unused3', 'unused4', 'metrics.icuCapacityRatio', 'riskLevels.overall', 
                   'metrics.vaccinationsInitiatedRatio', 'metrics.vaccinationsCompletedRatio', 
                   'actuals.vaccinesAdministered','riskLevels.caseDensity', 
                   'metrics.vaccinationsAdditionalDoseRatio', 'metrics.weeklyNewCasesPer100k', 'hsa', 
                   'hsaPopulation', 'metrics.bedsWithCovidPatientsRatio', 
                   'hsaName', 'metrics.weeklyCovidAdmissionsPer100k', 'communityLevels.cdcCommunityLevel', 
                   'communityLevels.canCommunityLevel', 'actuals.vaccinationsFall2022BivalentBooster',
                   'metrics.vaccinationsFall2022BivalentBoosterRatio']
```

```python
# Drop the columns
counties_dropped = counties_df.drop(columns_to_drop, axis=1)
counties_dropped
```

```python
# Verify count()
counties_dropped.count()
```

```python
# Drop any null values and verify columns lengths
counties_cleaned = counties_dropped.dropna()
counties_cleaned.count()
```

```python
# Save counties_cleaned
#counties_cleaned.to_csv("counties_cleaned.csv", sep=',', header=True, encoding='utf-8')
```

```python
# Get the completed dtypes for the ERD schema
counties_cleaned.dtypes
```

```python
# Creating a new DF with just the full columns
counties_dropped2 = counties_dropped[['date', 'state', 'county', 'fips', 'locationId', 'actuals.cases', 
                                      'actuals.deaths', 'actuals.newCases', 'actuals.newDeaths', 
                                      'cdcTransmissionLevel']]
counties_dropped2
```

```python
# Drop any null values and verify columns lengths
counties_cleaned2 = counties_dropped2.dropna()
counties_cleaned2.count()
```
