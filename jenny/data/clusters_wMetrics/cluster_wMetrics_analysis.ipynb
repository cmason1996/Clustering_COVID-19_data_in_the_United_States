```python
# Import segregated cluster tables

class_0_df = pd.read_csv('../clusters_wMetrics/class_0_wMetric.csv')
class_1_df = pd.read_csv('../clusters_wMetrics/class_1_wMetric.csv')
class_2_df = pd.read_csv('../clusters_wMetrics/class_2_wMetric.csv')
```

```python
# Generate pearson's correlation matrix and export to csv

class_0_corr_p = class_0_df.corr(method='pearson')
class_1_corr_p = class_1_df.corr(method='pearson')
class_2_corr_p = class_2_df.corr(method='pearson')

#export to csv
class_0_corr_p.to_csv('../clusters_wMetrics/class0_wMetric_pearson.csv')
class_1_corr_p.to_csv('../clusters_wMetrics/class1_wMetric_pearson.csv')
class_2_corr_p.to_csv('../clusters_wMetrics/class2_wMetric_pearson.csv')
```

```python
# condense the pearson's correlation matrix to an unstacked list with descending values 
class_0_pcorr = class_0_df.corr(method='pearson').unstack().sort_values(ascending=False).drop_duplicates()
class_0_pcorr.to_csv('../clusters_wMetrics/class0_pearson_unstacked.csv')
class_1_pcorr = class_1_df.corr(method='pearson').unstack().sort_values(ascending=False).drop_duplicates()
class_1_pcorr.to_csv('../clusters_wMetrics/class1_pearson_unstacked.csv')
class_2_pcorr = class_2_df.corr(method='pearson').unstack().sort_values(ascending=False).drop_duplicates()
class_2_pcorr.to_csv('../clusters_wMetrics/class2_pearson_unstacked.csv')
```

```python
# load pearson workbook sheets to dataframes
pearson_class0 = pd.read_excel("../jenny/data/pearson/pearson_summary.xlsx", sheet_name = 'class0_pearson_unstacked')
pearson_class1 = pd.read_excel("../jenny/data/pearson/pearson_summary.xlsx", sheet_name = 'class1_pearson_unstacked')
pearson_class2 = pd.read_excel("../jenny/data/pearson/pearson_summary.xlsx", sheet_name = 'class2_pearson_unstacked')
```

```python
pearson_class0_filtered = pearson_class0[pearson_class0['pearson'] >= 0.8]
pearson_class1_filtered = pearson_class1[pearson_class1['pearson'] >= 0.8]
pearson_class2_filtered = pearson_class2[pearson_class2['pearson'] >= 0.8]
```

```python
pearson_class0_filtered_neg = pearson_class0[pearson_class0['pearson'] <= -0.8]
pearson_class1_filtered_neg = pearson_class1[pearson_class1['pearson'] <= -0.8]
pearson_class2_filtered_neg = pearson_class2[pearson_class2['pearson'] <= -0.8]
```

```python
pearson_class2_filtered_neg
```

```python
pearson_list = [pearson_class0_filtered, pearson_class1_filtered, pearson_class2_filtered]
pearson_all = pd.concat(pearson_list)
pearson_all_dropValue = pearson_all.drop(columns='pearson')
pearson_all_drop = pearson_all_dropValue.drop_duplicates()
pearson_all_drop
```

```python
# drop all duplicate coorelation pairs from the pearson's dataframe. 
for index, row in pearson_all_drop.iterrows():
    if (pearson_all_drop['feature_1'] == row['feature_2']).all() & (pearson_all_drop['feature_2'] == row['feature_1']).all():
        pearson_all_drop.drop(index, inplace = True)

pearson_sharedFeatures = pearson_all_drop.reset_index()
pearson_sharedFeatures_clean = pearson_sharedFeatures.drop(columns='index')

# export pearson_sharedFeatures to csv
pearson_sharedFeatures_clean.to_csv('../jenny/data/pearson/pearson_sharedFeatures.csv')
```

```python
pearson_sharedFeatures_clean
```

```python
feature_1 = pearson_sharedFeatures_clean.iloc[:,0]
feature_2 = pearson_sharedFeatures_clean.iloc[:,1]
pearsons_stackedFeatures = pd.DataFrame(columns=['features'])
pearsons_stackedFeatures['features'] = pd.concat([feature_1, feature_2])
```

```python
# Create a dataframe of counts of each significant feature with the most amount of correlations
pearson_featureCount =pearsons_stackedFeatures['features'].value_counts().rename_axis('features').reset_index(name='counts')
pearson_featureCount
```

```python
ax = sns.barplot(data = pearson_featureCount, x = 'counts', y='features').set_title('Counts of features that have the most positive correlations')
```

```python
# heatmap showing just clustered #1  pearson correlation.
fig, ax = plt.subplots(figsize = ( 30 , 10 ))
dfCorr = class_0_df.corr()
filteredDf = dfCorr[((dfCorr >= .8) | (dfCorr <= -.8)) & (dfCorr !=1.000)]
sns.heatmap(filteredDf, annot=True, cmap="Reds")
ax.set_title('Heatmap of Cluster #0 Pearsons Correlation', size = 24)
plt.show()
```

```python
# condense the spearman's correlation matrix to an unstacked list with descending values 
class_0_scorr = class_0_df.corr(method='spearman').unstack().sort_values(ascending=False).drop_duplicates()
class_0_scorr.to_csv('../jenny/data/spearman/class0_spearman_unstacked.csv')
class_1_scorr = class_1_df.corr(method='spearman').unstack().sort_values(ascending=False).drop_duplicates()
class_1_scorr.to_csv('../jenny/data/spearman/class1_spearman_unstacked.csv')
class_2_scorr = class_2_df.corr(method='spearman').unstack().sort_values(ascending=False).drop_duplicates()
class_2_scorr.to_csv('../jenny/data/spearman/class2_spearman_unstacked.csv')
```
