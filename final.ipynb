```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import hvplot.pandas

from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('states_no_metrics_JMV.csv')
df
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
df = df.set_index('Unnamed: 0')
```

```python
df.head()
```

```python
# Separate this out of the features since it cannot be used for analysis, but want to keep for merging later
df_targets = df[['date', 'country', 'state', 'fips', 'locationId']].copy()
df_targets.head()
```

```python
# Separate the code to features to be used for the PCA
features_df = df.drop(columns = ['date', 'country', 'state', 'fips', 'locationId'])
features_df.dtypes
```

```python
# standardize the features using the StandardScaler Library
scaler = StandardScaler()
data_fit = scaler.fit(features_df)
X_scaled = data_fit.transform(features_df)
print(X_scaled[0:5])
```

```python
# Initialize the PCA model 
pca = PCA(n_components=3)

# Get three principle components for the data 
data_pca = pca.fit_transform(X_scaled)
data_pca
```

```python
# Create a dataframe for the principle components
pcs_df = pd.DataFrame(
    data=data_pca, columns=['PC 1', 'PC 2', 'PC 3'])
pcs_df
```

```python
# Create an elbow curve to find the best value for K.
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)
```

```python
# create a plot for the elbow curve:
elbow_data = {"k": k, "inertia" : inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
# Run k-means with k=3

# Initialize the K-Means model.
model = KMeans(n_clusters=3, random_state=0)

# Fit the model
model.fit(pcs_df)

# Predict clusters
predictions = model.predict(pcs_df)
predictions
```

```python
# Create a new DataFrame including predicted clusters and features.
# Concatentate the df_targets and pcs_df DataFrames on the same columns.
pcs_df.set_index(df_targets.index, inplace=True)
clustered_df = pd.concat([df_targets, pcs_df], axis = 1)

#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
clustered_df['Class'] = predictions

# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)
```

```python
# Creating a 3D-Scatter with the PCA data and the clusters
fig = px.scatter_3d(
    clustered_df,
    x="PC 1",
    y="PC 2",
    z="PC 3",
    hover_name='state',
    hover_data=['fips'],
    color="Class",
    symbol="Class",
    width=800,
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```

```python
# merge the class column to the original dataframe
merged_df = df.join(clustered_df['Class'])
merged_df
```

```python
# Separate the dataframe according to class

class_0_df = merged_df[(merged_df['Class'] == 0)]
class_1_df = merged_df[(merged_df['Class'] == 1)]
class_2_df = merged_df[(merged_df['Class'] == 2)]
```

```python
# export to csv
class_0_df.to_csv('class_0.csv', index=False)
class_1_df.to_csv('class_1.csv', index=False)
class_2_df.to_csv('class_2.csv', index=False)
```
